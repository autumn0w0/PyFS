WAP TO CHECK A LIST CONTAINS A PALINDROM OF ELEMENT


```python
l1 = [1, 2, 3, 4, 5]
l2 = [5, 4, 3, 2, 2]

cl1 = l1.copy()
cl1.reverse()

if(cl1 == l2):
    print("The lists are palindromes of each other")
else:
    print("The lists are not palindromes of each other")
```
