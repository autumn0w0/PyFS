Write a recursive function to calculate the sum of first n natural numbers.


```python
n = int(input("Enter a number: "))
def sum_of_n_recursive(n):
    if n == 0:
        return 0
    else:
        return n + sum_of_n_recursive(n - 1)

print(sum_of_n_recursive(n))
```

Write a recursive function to print all elements in a list.
Hint: use list & index as parameters.


```python
list1 = [1, 2, 3, 4, 5]

def print_list_recursive(list1, index=0):
    if index < len(list1):
        print(list1[index], end=' ')
        print_list_recursive(list1, index + 1)

print_list_recursive(list1)
```
