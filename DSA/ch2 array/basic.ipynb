```python
import requests
from bs4 import BeautifulSoup

def decode_secret_message(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Debug: print first 2000 chars of raw HTML to see structure
    print("=== RAW HTML SNIPPET ===")
    print(response.text[:3000])
    print("\n=== TABLES FOUND ===")
    tables = soup.find_all('table')
    print(f"Number of tables: {len(tables)}")
    
    print("\n=== ALL TR ROWS (first 5) ===")
    rows = soup.find_all('tr')
    print(f"Total rows: {len(rows)}")
    for i, row in enumerate(rows[:5]):
        print(f"Row {i}: {row}")

    print("\n=== TRYING p TAGS (first 10) ===")
    paras = soup.find_all('p')
    for i, p in enumerate(paras[:10]):
        print(f"p{i}: '{p.get_text(strip=True)}'")

url = "https://docs.google.com/document/d/e/2PACX-1vSvM5gDlNvt7npYHhp_XfsJvuntUhq1848y5xO_pA4b_gCWeXb6dM6ZxwN8rE6S4ghUsCj2VKR21oEP/pub"
decode_secret_message(url)
```
